```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Missing** **values**

```python
# Load the data
data = sns.load_dataset('titanic')
data
```

```python
# visualize the data
plt.figure(figsize= (8,5))
sns.heatmap(data.isnull(), cbar=False)
plt.show()
```

```python
data.info()
```

```python
data.isnull().sum().sort_values(ascending=False)
```

```python
round(data.isnull().sum() / len(data) *100, 2).sort_values(ascending=False)
```

```python
# calculate missing values
print("----------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum() / len(data) * 100, 2).sort_values(ascending=False)}")
```

```python
import missingno as msno

# load titanic dataset
data = sns.load_dataset('titanic')
# Visualize the data
msno.matrix(data,labels=True, fontsize=12, width_ratios=(2, 4), color=(0.2, 0.4, 0.6))
plt.show()
```

```python
data.head()
```

```python
round(data['age'].mean(), 2)
```

```python
data['age'].median()
```

```python
# Median of age to fill age missing values
data['age'] = data['age'].fillna(data['age'].median())
```

```python
data.drop('deck', axis=1, inplace=True)
```

```python
data.head()
```

```python
data['embark_town'].value_counts()
```

```python
# Replacing embark_town missing values with mode
data['embark_town'] = data['embark_town'].fillna(data['embark_town'].mode()[0])
data['embarked'] = data['embarked'].fillna(data['embarked'].mode()[0])
```

```python
data.isnull().sum().sort_values(ascending=False)
```

## **Forward** **and** **backward** **fill**

```python
df = sns.load_dataset('titanic')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# using forward fill impute age column
df['age']= df['age'].ffill()
```

```python
# using backward fill impute age column
df['age']= df['age'].bfill()
```
